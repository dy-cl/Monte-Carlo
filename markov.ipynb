```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

def psi(x, y):
    g1 = multivariate_normal([0.0, 0.0], [[2.0, 0], [0, 2.0]])
    g2 = multivariate_normal([7.0, 7.0], [[0.5, 0], [0, 5.0]])
    return g1.pdf(np.dstack((x, y))) + g2.pdf(np.dstack((x, y)))

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 15, 100)  
X, Y = np.meshgrid(x, y)
Z = psi(X, Y)

plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
```
