$$\frac{p_{circle}}{p_{total}} = \frac{A_{circle}}{A_{box}}$$ 

Where $p$ is the number of points, and $A$ is the area. Therefore, with $A_{box} = 1$ and $r_{circle} = 0.5$,

$$\frac{p_{circle}}{p_{total}} = \frac{\frac{1}{4}\pi}{1} = \frac{1}{4}\pi$$ 

Hence we can calculate $\pi$ by finding the ratio of points inside the circle to the total number (inside the box).

```python
import numpy as np
from matplotlib import pyplot as plt

#Set total number of points
num_points = 10000

#Initialise number of points inside the circle
p_circle = 0 

#Lists to store coordinates of points inside and outside the circle
points_inside_circle = []
points_outside_circle = []

#Loop over all random points
for i in range(num_points):

    #Generate random 
    x, y = np.random.uniform(0, 1, size = 2)

    #Find distance of point from the origin
    r = (x**2 + y**2)**0.5

    #If distance less than 1 (inside circle), increment p_circle and store the point coordinates
    if r < 1:
        p_circle += 1
        points_inside_circle.append((x, y))
    else:
        points_outside_circle.append((x, y))

#Calculate pi according to above formula
pi = 4*p_circle / num_points
print(pi)
```

In order to visualise this better, we can plot the points.

```python

#Extract x and y and plot points
plt.scatter(*zip(*points_inside_circle), color= 'green', s = 0.3)
plt.scatter(*zip(*points_outside_circle), color= 'grey', s = 0.3) 
plt.Circle((0, 0), 0.5)

#Plot unit circle
circle = plt.Circle((0, 0), 1, color='red', fill=False, label = 'Unit Circle')
plt.gca().add_patch(circle)

#Restrict plot limits to top right quartile of the circle
plt.xlim(0, 1)
plt.ylim(0, 1)
```
